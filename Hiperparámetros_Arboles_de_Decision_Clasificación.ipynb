```python
#Importamos librerias
import numpy as np
import pandas as pd
from sklearn import tree
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV, RandomizedSearchCV #Importamos librerias de partición
from sklearn.tree import DecisionTreeClassifier, export_graphviz #Libreria de arbol de decisiones
from scipy.stats import randint, uniform
```

```python
# Cargar el conjunto de datos en un DataFrame
df = pd.read_csv('/content/drive/MyDrive/Practica Diplomado/Entrenamiento de modelos/StressLevelDataset.csv')
df
```

```python
#Miramos la cantidad de estudiantes según el rango de estrés
estresEst = df['stress_level'].value_counts()
print("Estudiantes según el rango de estrés:\n",estresEst)
```

```python
#Creamos nuestro vector de variables independientes (caracteristicas)
x = df.drop(['stress_level'],axis = 1)
print("Caracteristicas originales:\n",x)
```

```python
#Creamos nuestro vector de etiquetas(stress_level)
y = df.stress_level
y
```

```python
#Creamos nuestros sets de entrenamiento
x_entre, x_test, y_entre, y_test = train_test_split(x, y, test_size=0.3, random_state=2024)
#En este caso vamos a utilizar el 30% de testeo y el 70% para entreno
```

```python
#Mostramos en pantalla nuestros sets de entrenamiento y testeo
print("Cantidad de datos para Tren de entreno: ", x_entre.shape, y_entre.shape)
print("Cantidad de datos para Tren de testeo: ", x_test.shape, y_test.shape)
```

```python
model = DecisionTreeClassifier()
```

#Validación cruzada

```python
scores = cross_val_score(model, x_entre, y_entre, cv=5, scoring='f1_macro')
print(scores)
```

```python
scores.mean()
```

#Hiperparámetros

```python
model.get_params()
```

```python
hiperparametros = {'criterion': ("gini", "entropy"),
              'min_samples_leaf': (1,2,3,4,5),
              'min_samples_split': (1,2,3,4,5),
              'max_depth' : (1,2,3,4,5)}
```

```python
rejilla = GridSearchCV(model,
                       hiperparametros,
                       scoring="accuracy")
```

```python
rejilla.fit(df[df.columns[:-1]].values,  #Se escoge todas las valiables menos la ultima
            df["stress_level"].values)   #Se escoge la variable objetivo
```

#Información de la Búsqueda en Rejilla

```python
sorted(rejilla.cv_results_.keys()) #Busca información potencial
```

#Rejilla de Parámetros Explorados


```python
rejilla.cv_results_["params"]
```

#Resultados de la Búsqueda en Rejilla

```python
print(rejilla.cv_results_["rank_test_score"])
print(rejilla.cv_results_["mean_test_score"])
print(rejilla.best_score_)
print(rejilla.best_params_)
```
